```python
networks1 = MLE1(input_size=784, hidden_size_list=[100, 100, 100, 100, 100], output_size=10,
                                weight_init_std=0.293)
networks2 = MLE2(input_size=784, hidden_size_list=[100, 100, 100, 100, 100], output_size=10,
                                weight_init_std=0.293)
```

```python
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True)

# 학습 데이터를 줄임
x_train = x_train[:1000]
t_train = t_train[:1000]

max_epochs = 20
train_size = x_train.shape[0]
batch_size = 100
learning_rate = 0.01
```

```python
batch_mask = np.random.choice(train_size, batch_size)
x_batch = x_train[batch_mask]
t_batch = t_train[batch_mask]

pred1 = networks1.predict(x_batch)
pred2 = networks2.predict(x_batch)
y1 = networks1.last_layer.forward(pred1, t_batch)
y2 = networks2.last_layer.forward(pred1, t_batch)
dy1 = networks1.last_layer.backward(1)
dy2 = networks2.last_layer.backward(1)

grads1 = networks1.gradient(x_batch, t_batch)
grads2 = networks2.gradient(x_batch, t_batch)
```
